```python
import pandas as pd
import matplotlib.pyplot as plt

root = "BT_res50_64_100_1/"

# Read the CSV files
df_std = pd.read_csv(root + 'default_representation_std.csv')
df_loss = pd.read_csv(root + 'default_train_loss.csv')
df_accuracy = pd.read_csv(root + 'default_kNN_accuracy.csv')

# Create main figure and axis with larger size
fig, ax1 = plt.subplots(figsize=(10, 6))

# Set scientific notation for y-axis
ax1.ticklabel_format(style='sci', axis='y', scilimits=(0,0))

# Plot Standard Deviation
color = 'tab:blue'
ax1.set_xlabel('Step')
ax1.set_ylabel('Standard Deviation', color=color)
line1, = ax1.plot(df_std['step'], df_std['representation_std'], color=color, label='Standard Deviation')
ax1.tick_params(axis='y', labelcolor=color)
ax1.grid(True)

# Create second y axis for Train Loss
ax2 = ax1.twinx()
ax2.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
color = 'tab:red'
ax2.set_ylabel('Train Loss', color=color)
line2, = ax2.plot(df_loss['step'], df_loss['train_loss'], color=color, label='Train Loss')
ax2.tick_params(axis='y', labelcolor=color)

# Create third y axis for kNN Accuracy
ax3 = ax1.twinx()
ax3.spines['right'].set_position(('outward', 60))
ax3.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
color = 'tab:green'
ax3.set_ylabel('kNN Accuracy', color=color)
line3, = ax3.plot(df_accuracy['step'], df_accuracy['kNN_accuracy'], color=color, label='kNN Accuracy')
ax3.tick_params(axis='y', labelcolor=color)

# Add legend
fig.legend(handles=[line1, line2, line3], loc='upper left')

plt.title('Standard Deviation, Train Loss and kNN Accuracy over Steps')

# Show the plots
fig.tight_layout()
plt.show()
```

**Whats**
- STD (better good)
- Loss (lower good)
- Acc (better Good)   

**Findings**.

- Strong correlations between STD, accuracy, and Loss function, 
- Model do not collpase (KNN converge)

**Questions**.
- Model have a worse representation(STD), better acc, 

# 2 We Compare different model to check (overfit or not)

```python
import pandas as pd
import matplotlib.pyplot as plt
import glob
import re
import os
import numpy as np 
# Find all directories matching pattern
# Find all directories matching pattern
dirs = glob.glob("BT_res50_64_100_*")
dirs = [dir for dir in dirs if os.path.isdir(dir)]

# Prepare a color cycle for plot lines
colors = plt.cm.viridis(np.linspace(0, 1, len(dirs)))

# Create main figure and axis
fig, axes = plt.subplots(3, 1, sharex=True, figsize=(10, 15))
fig.suptitle('Comparison of Standard Deviation, Train Loss and kNN Accuracy over Epochs')

for i, dir in enumerate(dirs):
    # Extract value from directory name
    value = re.search(r'BT_res50_64_100_(.*)', dir).group(1)
    
    # Read the CSV files
    df_std = pd.read_csv(dir + '/default_representation_std.csv')
    df_loss = pd.read_csv(dir + '/default_train_loss.csv')
    df_accuracy = pd.read_csv(dir + '/default_kNN_accuracy.csv')
    df_others = pd.read_csv(dir + '/default_others.csv')
    
    # Merge the dataframes on 'step'
    df_std = df_std.merge(df_others, on='step', how='inner')
    df_loss = df_loss.merge(df_others, on='step', how='inner')
    df_accuracy = df_accuracy.merge(df_others, on='step', how='inner')
    
    # Group by 'epoch' and calculate mean
    df_std = df_std.groupby('epoch').mean().reset_index()
    df_loss = df_loss.groupby('epoch').mean().reset_index()
    df_accuracy = df_accuracy.groupby('epoch').mean().reset_index()
    
    # Plot Standard Deviation
    axes[0].plot(df_std['epoch'], df_std['representation_std'], color=colors[i], label=f'@ {value}')
    axes[0].set_ylabel('Standard Deviation')
    axes[0].grid(True)
    
    # Plot Train Loss
    axes[1].plot(df_loss['epoch'], df_loss['train_loss'], color=colors[i], label=f'@ {value}')
    axes[1].set_ylabel('Train Loss')
    axes[1].grid(True)
    
    # Plot kNN Accuracy
    axes[2].plot(df_accuracy['epoch'], df_accuracy['kNN_accuracy'], color=colors[i], label=f'@ {value}')
    axes[2].set_xlabel('Epoch')
    axes[2].set_ylabel('kNN Accuracy')
    axes[2].grid(True)

# Add legend
for ax in axes:
    ax.legend(loc='best')
    ax.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
    ax.set_xlim([df_others['epoch'].min(), df_others['epoch'].max()])  # Setting x-axis limit

# Show the plots
plt.tight_layout()
plt.show()

```

**Whats**
- @ different part datasets


**Findings**.

- Bigger dataset, better performance roughly 
- Bottleneck is 60 epoch for current configuration

**Questions**.
- Model have a worse representation(STD), and model stop training suggests
- why this happened,

# Result looks good but for analysis Check the distribution

```python
import os
import glob
import matplotlib.pyplot as plt
import seaborn as sns

# Set a consistent style for seaborn
sns.set_theme(style="whitegrid")

# Path to the training and testing data
data_dir_train = 'C:/Users/isxzl/OneDrive/Code/AutoSSL/Datasets/xBD/train'
data_dir_test = 'C:/Users/isxzl/OneDrive/Code/AutoSSL/Datasets/xBD/test'

# Class names
classes = ['destroyed', 'major-damage', 'minor-damage', 'no_building', 'no-damage']

# Count the number of images for each class
image_counts_train = {class_label: len(glob.glob(os.path.join(data_dir_train, class_label, '*.png'))) for class_label in classes}
image_counts_test = {class_label: len(glob.glob(os.path.join(data_dir_test, class_label, '*.png'))) for class_label in classes}

# Create a figure and axes
fig, ax = plt.subplots(figsize=(12, 8))

# Define bar width
bar_width = 0.35

# Create the bar plot for training data
bars_train = plt.bar(np.arange(len(classes)) - bar_width/2, image_counts_train.values(), bar_width, color=sns.color_palette("Paired", len(classes)))

# Create the bar plot for testing data
bars_test = plt.bar(np.arange(len(classes)) + bar_width/2, image_counts_test.values(), bar_width, color=sns.color_palette("dark", len(classes)))

# Add a title and labels with larger fonts
plt.title('Distribution of Images Across Different Classes in the Training and Testing Datasets', fontsize=18)
plt.xlabel('Class Label', fontsize=14)
plt.ylabel('Number of Images', fontsize=14)

# Increase the size of the tick labels
plt.xticks(np.arange(len(classes)), image_counts_train.keys(), fontsize=12, rotation=45)
plt.yticks(fontsize=12)

# Add a legend
plt.legend([bars_train, bars_test], ["Training", "Testing"], title="Dataset", title_fontsize='13', fontsize='12')

# Show the plot
plt.show()
```

**Findings**.
- We have unbalanced data

# Now we could dip into the confusion matrix

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
# Create a mapping of numerical class labels to their corresponding names
class_label_map = {
    0: 'destroyed',
    1: 'major-damage',
    2: 'minor-damage',
    3: 'no_building',
    4: 'no-damage'
}

def plot_confusion_matrix(excel_file):
    # Load Excel file
    xls = pd.ExcelFile(excel_file)
    
    # Get all sheet names
    sheet_names = xls.sheet_names

    for sheet in sheet_names:
        # Load sheet into a DataFrame
        df = pd.read_excel(xls, sheet, index_col=0)
        
        # Replace numerical labels with names
        df = df.rename(columns=class_label_map, index=class_label_map)

        plt.figure(figsize=(10, 7))
        
        # Display confusion matrix
        sns.heatmap(df, annot=True, cmap='Blues', fmt=".2f")

        plt.title('Confusion Matrix for ' + sheet)
        plt.xlabel('Predicted')
        plt.ylabel('Actual')

        # Set the labels of the x and y axis to the class names
        plt.xticks(ticks=np.arange(0.5, len(df.columns), 1), labels=df.columns, rotation=45)
        plt.yticks(ticks=np.arange(0.5, len(df.index), 1), labels=df.index, rotation=0)

        plt.show()

plot_confusion_matrix('^BT_res18_64_100_.+$_confusion.xlsx')
```
